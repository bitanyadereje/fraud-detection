```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
credit = pd.read_csv('../data/raw/creditcard.csv')
print(credit.shape)
credit.head()
```

```python
print(credit.isnull().sum().sum())  
credit['Class'].value_counts(normalize=True)
```

```python
print(credit.columns.tolist())
```

```python
from sklearn.model_selection import train_test_split

X = credit.drop('Class', axis=1)
y = credit['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print(f"Train fraud ratio: {y_train.mean():.5f}")
print(f"Test fraud ratio: {y_test.mean():.5f}")

# Save
X_train.to_csv('../data/processed/credit_X_train.csv', index=False)
X_test.to_csv('../data/processed/credit_X_test.csv', index=False)
y_train.to_csv('../data/processed/credit_y_train.csv', index=False)
y_test.to_csv('../data/processed/credit_y_test.csv', index=False)
print("Saved credit card processed datasets.")
```
